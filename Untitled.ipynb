```python
#50km or 50mi
#df[(df['Event distance/length'] == '50km') | (df['Event distance/length'] == '50mi')]
df[df['Event distance/length'].isin(['50km','50mi'])]
```

```python
#Marathons in USA
df['Event Country']=df['Event name'].str.split('(').str.get(1)
#df['Event country']=df['Event name'].str.split('(')
```

```python
df.columns
```

```python
df.head()
```

```python
#df['Event Country'] = df['Event Country'].str[:-1]
df[df['Event Country']=='USA']
```

```python
# i need tp combine my 3 filters to get the data for marathons in USA, 50KM and 50KI for 2020
#df[df['Year of event']==2020]
#df[df['Event distance/length'].isin(['50km','50mi'])]
#df[df['Event Country']=='USA']
df2=df[(df['Year of event']==2020)  & (df['Event distance/length'].isin(['50km','50mi']))  &  (df['Event Country']=='USA') ]
```

```python
df2.head()
```

```python
df2.shape
```

```python
#remoev (USA) from Event name
df2['Event name'] = df2['Event name'].str[:-5]
```

```python
df2.head()
```

```python
#clean up athelete age
df2['athlete_age']=2020 - df2['Athlete year of birth']
```

```python
#remove h from 'Athlete performance' column
df2['Athlete performance']=df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
#drop some columns:
df3=df2[['Year of event','Event dates','Event name','Event distance/length','Event number of finishers','Athlete performance','Athlete gender','Athlete average speed','Athlete ID','athlete_age']]
```

```python
df3.head()
```

```python
#clean up null values
df3.dropna(inplace=True)
```

```python
#check for duplicated values
df3[df3.duplicated()==True]
```

```python
#reset index
df3.reset_index(drop=True)
```

```python
#fixe types
df3.dtypes
```

```python
df3['Athlete average speed']=df3['Athlete average speed'].astype(float)
```

```python
df3['athlete_age']=df3['athlete_age'].astype(int)
```

```python
df3.dtypes
```

```python
df3.head()
```

```python
#rename columns
df3=df3.rename(columns={'Year of event':'year' ,
                       'Event dates':'race_day' ,
                       'Event name':'race_name' ,
                       'Event distance/length':'race_lenght' ,
                       'Event number of finishers':'race_number_of_finishers' ,
                       'Athlete performance':'athlete_performance' ,
                       'Athlete gender':'athlete_gender' ,
                       'Athlete average speed':'athlete_average_speed' ,
                       'Athlete ID':'athlete_id' })
```

```python
df3.head()
```

```python
df3.info()
```

```python
#reorganization of columns
df3=df3[['race_day','race_name','race_lenght','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed','year']]
```
